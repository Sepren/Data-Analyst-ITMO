````python
def remove_hyphens(value):
    """
    Удаляет символы '-' из строки.
    Если значение не строка, возвращает его без изменений.
    """
    if pd.isna(value):  # Проверка на NaN
        return value
    return str(value).replace('-', '')  # Удаляем все '-'
````

````python
def remove_parentheses(value):
    """
    Удаляет только открывающие и закрывающие скобки из строки.
    Если значение не строка, возвращает его без изменений.
    """
    if pd.isna(value):  # Проверка на NaN
        return value
    return re.sub(r'[()]', '', str(value))  # Удаляем только '(' и ')'
````

````python
df.loc[:, 'phone_email'] = df['phone_email'].apply(remove_parentheses)
````

````python
def normalize_11_digit_phone(digits):
    """
    Обрабатывает 11-значные номера телефонов.
    """
    if len(digits) == 11:
        if digits.startswith('7'):
            return digits
        elif digits.startswith('8'):
            return '7' + digits[1:]
    return None
````

````python
def normalize_10_digit_phone(digits):
    """
    Обрабатывает 10-значные номера телефонов (без кода страны).
    """
    if len(digits) == 10:
        return '7' + digits
    return None
````

````python
def normalize_7_digit_phone(digits):
    """
    Обрабатывает 7-значные номера телефонов (городские).
    """
    if len(digits) == 7:
        return '7812' + digits  # Добавляем код города (например, 812 для Санкт-Петербурга)
    return None
````

````python
def normalize_8_digit_phone(digits):
    """
    Обрабатывает 8-значные номера телефонов.
    """
    if len(digits) == 8:
        return '781' + digits  # Добавляем 3 цифры в самое начало
    return None
````

````python
def filter_9_digit_phone(digits):
    """
    Удаляем 9-значные номера телефонов.
    """
    if pd.isna(digits):  # Проверка на NaN
        return True  # Сохраняем строки с NaN
    return len(str(digits)) != 9  # Удаляем строки с 9 символами
````

````python
df = df.loc[df['phone_email'].apply(filter_9_digit_phone)]
````

````python
def normalize_12_digit_phone(digits):
    """
    Обрабатывает 12-значные номера телефонов.
    """
    if len(digits) == 12 and digits.startswith('989') or digits.startswith('789'):
        return '7' + digits[2:]
    elif len(digits) == 12 and digits.startswith('89') or digits.startswith('8495'):
        return '7' + digits[1:-1]
    return None
````

````python
def filter_12_digit_phone(digits):
    """
    Удаляем остальные 12-значные номера телефонов.
    """
    if pd.isna(digits):  # Проверка на NaN
        return True  # Сохраняем строки с NaN
    return len(str(digits)) != 12  # Удаляем строки с 12 символами
````

````python
df = df.loc[df['phone_email'].apply(filter_12_digit_phone)]
````

````python
def filter_short_entries(value):
    """
    Фильтрует строки, длина которых составляет меньше 4 символа.
    Возвращает True для строк, которые остаются, и False для тех, которые нужно удалить.
    """
    if pd.isna(value):  # Проверка на NaN
        return True  # Сохраняем строки с NaN
    return len(str(value)) > 3  # Удаляем строки с длиной меньше 3 символов
````

````python
df = df.loc[df['phone_email'].apply(filter_short_entries)]
````

````python
def normalize_13_digit_phone(digits):
    """
    Обрабатывает 13-значные номера телефонов.
    """
    if len(digits) == 13 and digits.startswith('989') or digits.startswith('789'):
        return '7' + digits[2:]
    elif len(digits) == 13 and digits.startswith('89') or digits.startswith('8495'):
        return '7' + digits[1:-1]
    elif len(digits) == 13 and digits.startswith('79') or digits.startswith('8495'):
        return digits[:-2]
    elif len(digits) == 13 and digits.startswith('+79'):
        return digits[1:-1]
    return None
````

````python
def filter_13_digit_phone(digits):
    """
    Удаляем остальные 13-значные номера телефонов.
    """
    if pd.isna(digits):  # Проверка на NaN
        return True  # Сохраняем строки с NaN
    return len(str(digits)) != 13  # Удаляем строки с 13 символами
````

````python
df = df.loc[df['phone_email'].apply(filter_13_digit_phone)]
````

````python
def filter_16_digit_phone(digits):
    """
    Удаляем остальные 16-значные номера телефонов.
    """
    if pd.isna(digits):  # Проверка на NaN
        return True  # Сохраняем строки с NaN
    return len(str(digits)) != 16  # Удаляем строки с 16 символами
````

````python
df = df.loc[df['phone_email'].apply(filter_16_digit_phone)]
````

````python
def extract_numbers(value):
    """
    Извлекает номера телефонов из строки, разделяя их по разным разделителям.
    Убирает очевидные некорректные части текста.
    """
    if pd.isna(value):
        return []

    # Разделяем строку на части по разделителям и выделяем только числовые последовательности
    parts = re.split(r'[,\s/\\]+', str(value))

    # Убираем части, которые содержат нецифровые символы, кроме корректных номеров
    numbers = [
        re.sub(r'\D', '', part)  # Убираем все, кроме цифр
        for part in parts if re.search(r'\d', part)  # Оставляем только части с цифрами
    ]

    # Отфильтровываем некорректные номера
    numbers = [normalize_phone(num) for num in numbers if num.strip()]
    return numbers
````

````python
def normalize_phone(phone_email):
    """
    Основная функция нормализации номера телефона.
    Вызывает специализированные функции для обработки.
    Если номер невалидный, возвращает исходное значение.
    """
    digits = extract_numbers(phone_email)
    if not digits or len(digits) < 7:  # Невалидные данные (меньше 7 цифр)
        return phone_email  # Возвращаем исходное значение

    # Проверяем различные форматы
    normalized = (
        normalize_11_digit_phone(digits) or
        normalize_10_digit_phone(digits) or
        normalize_7_digit_phone(digits) or
        normalize_12_digit_phone(digits) or
        normalize_8_digit_phone(digits) or
        normalize_13_digit_phone(digits)
    )

    return normalized or phone_email  # Возвращаем нормализованный номер или исходное значение
````

````python
def split_and_classify_phones(value):
    """
    Обрабатывает строку, выделяя основной номер и дополнительные.
    """
    numbers = extract_numbers(value)

    if not numbers:  # Если номера не найдены
        return None, []

    # Считаем первый номер основным, а остальные дополнительными
    return numbers[0], numbers[1:]
````
