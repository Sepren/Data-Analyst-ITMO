````python
def remove_hyphens(value):
    """
    Удаляет символы '-' из строки.
    Если значение не строка, возвращает его без изменений.
    """
    if pd.isna(value):  # Проверка на NaN
        return value
    if len(str(value).replace('-', '')) != len(value):
        print(value)
    return str(value).replace('-', '')  # Удаляем все '-'
````

````python
def remove_parentheses(value):
    """
    Удаляет только открывающие и закрывающие скобки из строки.
    Если значение не строка, возвращает его без изменений.
    """
    if pd.isna(value):  # Проверка на NaN
        return value
    if len(re.sub(r'[()]', '', str(value))) != len(value):
        print(value)
    return re.sub(r'[()]', '', str(value))  # Удаляем только '(' и ')'
````

````python
def normalize_12_digit_phone(digits):
    """
    Обрабатывает 12-значные номера телефонов.
    """
    if digits is None or pd.isna(digits):  # Проверяем, что значение не None или NaN
        return digits

    digits = str(digits)  # Преобразуем в строку на случай, если это число
    if len(digits) == 12:
        print(f"Обрабатывается: {digits}")
        if digits[:3] == '989' or digits[:3] == '789':
            print(f"Условие сработало для: {digits[:3]}")
            return '7' + digits[2:]
        elif digits[:2] == '89' or digits[:4] == '8495' or digits[:2] == '99':
            print(f"Условие сработало для: {digits[:2]} или {digits[:4]}")
            return '7' + digits[1:-1]
        elif digits[:2] == '79':
            print(f"Условие сработало для: {digits[:2]}")
            return digits[:-1]
        else:
            print(f"Никакие условия не подошли для: {digits}")


    return None
````

````python
def filter_short_entries(value):
    """
    Фильтрует строки, длина которых составляет меньше 4 символа.
    Возвращает True для строк, которые остаются, и False для тех, которые нужно удалить.
    """
    if pd.isna(value):  # Проверка на NaN
        return True  # Сохраняем строки с NaN
    if len(str(value)) < 4:
        print(value)
    return len(str(value)) > 3  # Удаляем строки с длиной меньше 3 символов
````

````python
def normalize_13_digit_phone(digits):
    """
    Обрабатывает 13-значные номера телефонов.
    """
    if digits is None or pd.isna(digits):  # Проверяем, что значение не None или NaN
        return digits

    digits = str(digits)  # Преобразуем в строку на случай, если это число
    # print(f"Обрабатывается: {digits}")

    if len(digits) == 13:
        print(f"Обрабатывается: {digits}")
        # Если начинается с 989 или 789, обрезаем с 3-го символа и добавляем '7'
        if digits[:3] in ['989', '789']:
            print(f"Совпало с условием 989/789: {digits} -> {'7' + digits[2:-1]}")
            return '7' + digits[2:-1]

        # Если начинается с 89 или 8495, заменяем на номер с 7
        elif digits[:2] == '89' or digits[:4] == '8495':
            print(f"Совпало с условием 89/8495: {digits} -> {'7' + digits[1:-2]}")
            return '7' + digits[1:-2]

        # Если начинается с 79, убираем последние 2 символа
        elif digits[:2] == '79':
            print(f"Совпало с условием 79: {digits} -> {digits[:-3]}")
            return digits[:-3]

        # Если начинается с +79, удаляем плюс и последние 2 символа
        elif digits[:3] == '+79':
            print(f"Совпало с условием +79: {digits} -> {digits[1:-2]}")
            return digits[1:-2]

        else:
            print(f"Никакие условия не подошли для: {digits}")

    return None  # Возвращаем номер без изменений, если ни одно условие не подошло
````

````python
def process_and_filter_phone_spaces_v4(phone_column):
    """
    Обрабатывает строки, удаляя пробелы внутри номеров и исключая блоки менее 5 цифр.
    Возвращает изменённые строки: оригинальные строки, строки после промежуточной обработки и конечный результат.
    """
    original_strings = []  # Оригинальные строки
    processed_strings = []  # Строки после промежуточной обработки
    final_results = []  # Конечный результат

    for value in phone_column:
        if pd.isna(value):  # Пропускаем NaN
            continue

        # Разделяем строку по основным разделителям
        parts = re.split(r'[,\s/]+', value.strip())  # Пробелы, запятые, слэши
        numbers = []
        temp_number = ""

        for part in parts:
            clean_part = re.sub(r'\D', '', part)  # Убираем нецифровые символы
            if len(temp_number) + len(clean_part) <= 11:
                # Если номер не превышает 11 символов, объединяем
                temp_number += clean_part
            else:
                # Добавляем законченный номер в список
                if temp_number:
                    numbers.append(temp_number)
                temp_number = clean_part  # Начинаем новый номер

        # Добавляем последний собранный номер
        if temp_number:
            numbers.append(temp_number)

        # Собираем строку обратно с разделителем
        processed_string = " ".join(numbers)

        # Фильтрация коротких блоков (менее 5 символов)
        filtered_numbers = [num for num in numbers if len(num) >= 5]

        # Если после фильтрации остались данные, добавляем их как конечный результат
        final_result = " ".join(filtered_numbers) if filtered_numbers else None

        # Добавляем только изменённые строки
        if value != processed_string:  # Условие изменения
            original_strings.append(value)
            processed_strings.append(processed_string)
            final_results.append(final_result)

    return original_strings, processed_strings, final_results
````

````python
def split_and_process_phones(value):
    """
    Разделяет строку на несколько номеров по различным разделителям,
    обрабатывает каждый номер и возвращает список.
    """
    if pd.isna(value):
        return [None]
    # Разделяем строку по любым разделителям (запятая, слэш, пробелы и т.д.)
    parts = re.split(r'[,\s/\\]+', str(value))
    # Убираем пустые строки и обрабатываем номера
    processed_parts = [normalize_phone(remove_parentheses(part)) for part in parts if part.strip()]
    return processed_parts
````

````python
def split_and_classify_phones(value):
    """
    Разделяет строку на отдельные номера телефонов и классифицирует их.
    """
    if pd.isna(value):
        return []

    # Разделяем строку на части по различным разделителям
    parts = re.split(r'[,\s/\\]+', str(value))

    # Удаляем лишние символы и нормализуем номера
    numbers = [re.sub(r'\D', '', part) for part in parts if re.search(r'\d', part)]

    # Обрабатываем номера, используя функции нормализации
    processed_numbers = [normalize_phone(num) for num in numbers if num.strip()]

    return processed_numbers
````

````python
def extract_numbers(value):
    """
    Извлекает номера телефонов из строки, разделяя их по разным разделителям.
    Убирает очевидные некорректные части текста.
    """
    if pd.isna(value):
        return []

    # Разделяем строку на части по разделителям и выделяем только числовые последовательности
    parts = re.split(r'[,\s/\\]+', str(value))

    # Убираем части, которые содержат нецифровые символы, кроме корректных номеров
    numbers = [
        re.sub(r'\D', '', part)  # Убираем все, кроме цифр
        for part in parts if re.search(r'\d', part)  # Оставляем только части с цифрами
    ]

    # Отфильтровываем некорректные номера
    numbers = [normalize_phone(num) for num in numbers if num.strip()]
    return numbers
````

````python
def split_and_classify_phones(value):
    """
    Обрабатывает строку, выделяя основной номер и дополнительные.
    """
    numbers = extract_numbers(value)

    if not numbers:  # Если номера не найдены
        return None, []

    # Считаем первый номер основным, а остальные дополнительными
    return numbers[0], numbers[1:]
````
